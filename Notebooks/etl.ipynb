#  DF Ventas

```python
df_ventas = pd.read_csv('Ventas.csv', sep= ';', low_memory=False)
```

```python
df_ventas
```

```python
# Verifico nulos 
df_ventas.info()
```

### parece que hay una gran cantidad de nulos  

```python
# elimino todo los nuelos  
df_ventas = df_ventas.dropna()
```

```python
# verifico
df_ventas.info()
```

```python
# cambio el tipo de datoa entero 
df_ventas.Empleado = df_ventas.Empleado.astype(int)
```

```python
# verifico
df_ventas.info()
```

```python
# cambio el tipo de dato a entero   
df_ventas.CodigoFamilia = df_ventas.CodigoFamilia.astype(int)
df_ventas.Area = df_ventas.Area.astype(int)
```

```python
# Convertir la columna 'Fecha' de tipo object a tipo fecha con formato "día mes año"
df_ventas['Fecha'] = pd.to_datetime(df_ventas['Fecha'], format='%d/%m/%Y')
```

```python
# Limpia los valores en la columna 'Ventas'
df_ventas['Ventas'] = df_ventas['Ventas'].str.replace('.', '').str.replace(',', '.')
```

```python
# cambio el tiepo de datios a Float 
df_ventas['Ventas'] = df_ventas['Ventas'].astype(float)
```

```python
# Limpia los valores en la columna 'Cantidad'
df_ventas['Cantidad'] = df_ventas['Cantidad'].str.replace(',','')
```

```python
# cambio valoeres de la columna Cantidad a entero
df_ventas['Cantidad'] = df_ventas['Cantidad'].astype(int)
```

```python
df_ventas.info()
```

```python
df_ventas.tail()
```

```python
# Exporto a un nuevo csv

df_ventas.to_csv('Ventas_ETL.csv', index=False)
```

# EDA   Ventas

```python
# cantidad de empleados por sucursal 
df_ventas.groupby('Sede')['Empleado'].nunique().sort_values(ascending=False)
```

```python
# cantida de ventas por empleado 
df_ventas['Empleado'].value_counts()
```

```python
# cantida de ventas por localidad
df_ventas['Localidad'].value_counts()
```

```python
# Extrae el mes y el año de la columna 'Fecha'
df_ventas['Año'] = df_ventas['Fecha'].dt.year
df_ventas['Mes'] = df_ventas['Fecha'].dt.month

# Agrupa las ventas por mes y año y calcula la suma de ventas para cada combinación de mes y año
ventas_por_mes_y_año = df_ventas.groupby(['Año', 'Mes'])['Ventas'].sum()

# Grafica las ventas por mes y año con líneas continuas
fig, ax = plt.subplots(figsize=(12, 6))
ventas_por_mes_y_año.plot(marker='o', linestyle='-', ax=ax)
plt.xlabel('Año - Mes')
plt.ylabel('Ventas')
plt.title('Ventas por Mes y Año')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de dispersión para el campo 'Ventas'
plt.figure(figsize=(8, 6))
plt.scatter(range(len(df_ventas)), df_ventas['Ventas'], marker='o', color='b', label='Datos')
outliers_precio = df_ventas[df_ventas['Ventas'] > df_ventas['Ventas'].quantile(0.95)]
plt.scatter(outliers_precio.index, outliers_precio['Ventas'], marker='x', color='r', label='Outliers')
plt.title('Gráfico de Dispersión - Precio de venta')
plt.xlabel('Índice')
plt.ylabel('Precio de venta')
plt.legend()
plt.show()
```

```python
# Gráfico de dispersión para el campo 'Cantidad'
plt.figure(figsize=(8, 6))
plt.scatter(range(len(df_ventas)), df_ventas['Cantidad'], marker='o', color='b', label='Datos')
outliers_cantidad = df_ventas[df_ventas['Cantidad'] > df_ventas['Cantidad'].quantile(0.95)]
plt.scatter(outliers_cantidad.index, outliers_cantidad['Cantidad'], marker='x', color='r', label='Outliers')
plt.title('Gráfico de Dispersión - Cantidad')
plt.xlabel('Índice')
plt.ylabel('Cantidad')
plt.legend()
plt.show()
```
